## Prompt Experiments

### 1. Knowing Skills Based on Gymnastics Levels

I want to see what the model currently knows about levels to see if this needs to be fine tuned or if it knows this information already

```python
levels = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[
        {"role": "user", 
         "content": """
         Act as a gymnastics coach.  List 10 skills needed to participate in each level and 10 skills that the gymnasts are 
         working on for the next level.  Provide break downs for beginner, advanced beginner, intermediate, and advanced 
         recreational classes; all xcel levels; and all junior olympics developmental levels.
         """}
    ]
)
```

```python
levels_content = levels.choices[0].message.content
print(f"{levels_content}")
```

#### Review of response:
The model does have some basic understanding of skill prgression with levels and a general understanding of what skills are needed for each level.  There are some skills that I see listed under the wrong levels, so this will require some fine tuning.

### 2. List drills for a specific skill

Lets see what the LLM can create for drills.  I'm going to ask about one of the most fundamental gymnastics skills that takes a long time to perfect - the back handspring.  I'm going to ask for several drills for each level and I want it to provide several details for each drill.

```python
BHS_drills = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[
        {"role": "user", 
         "content": """
         Provide 3 back handspring drills for each: beginners, advanced beginners, intermediate, and advanced recreational classes.
         Provide a detailed description on how to complete the drill, what part of the back handspring the drill is working on,
         what equipment is necessary to complete the drill, etc.
         """}
    ]
)
```

```python
BHS_drills_content = BHS_drills.choices[0].message.content
print(f"{BHS_drills_content}")
```

#### Review of response:
The model has a fairly good understanding of back handspring progression for each level.  There are some drills that I see listed under the wrong levels, so this will require some fine tuning. I would like to see a little more description on the equipment - does this need a 4", 8", 12" or 16" mat? what size and firmness of block or incline, etc. I think the description of the drill could also be a little more detailed.  Most of them, however, give me a good idea of what the drill is and what it is trying to accomplish.

### 3. Event Practice Plan
The goal is for it to create a practice plan, so lets see how it does with a single event rotation.

```python
bars_practice = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[
        {"role": "user", 
         "content": """
         Create a practice plan for intermediate girls bars. They have 30 minutes on the event.  
         They need 5 minutes for warm up - this should include some light conditioning. 
         They need 5-6 stations that they will complete over the remaining 25 minutes.
         Provide a detailed description of each station. 
         If it is a drill, provide details on what skill and what part of the skill the drill is working on.
         If equipment other than the bar is needed, list the equipment that is needed.
         """}
    ]
)
```

```python
bars_practice_content = bars_practice.choices[0].message.content
print(f"{bars_practice_content}")
```

#### Review of response:
This one could definitely use some more fine tuning.  Giants and glide swings are very different levels of skills. Also,  - swing to handstand isn't really a thing - there's cast to handstands and giants (which are swinging all the way around the bar and can stop in a handstand).  Cast handstand and giants are both too high of skills for an intermediate recreational class.  I would also like to see an aparatus specific warm up and conditioning list - i.e. chin ups, leg lifts, swings, etc.

### 5. Full class plan
The number of events will vary by gym, so this will have to be kept in mind while developing the app.  I've been to gyms that have 3 hour practices with 30 minutes of warm up and stretching, an hour at one event, an hour at another event, and then 30 minutes of conditioning.  I've also been to ones that have 2 hour practices with no significant warm up and stretching (5 minutes built into first event), 30 minutes at each of the 4 events, no dedicated conditioning time, and 5 minutes of fun activities built into the time on the last event.

```python
adv_rec = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[
        {"role": "user", 
         "content": """
         Create a gymnastics practice plan for my advanced recreational class. It is a 2 hour practice. 
         30 minutes at each of the following events in order: trampoline, floor, beam, and uneven bars
         the first 5 minutes of each event should include event specific warm ups and light conditioning
         the warm up of the first event should also include light stretching
         I need 5 stations for each event - provide detailed descriptions and equipment needed for each station
         if station is a drill, provide description of the skill the drill is for
         allow 5 minutes of the last event to be for active play - choose the activity
          """}
    ]
)
```

```python
adv_rec_content = adv_rec.choices[0].message.content
print(f"{adv_rec_content}")
```

#### Review of response:
Overall, this is a great start to a practice plan.  some of the stations are too simple and some are too complex, but that has been consistant with the model not quite knowing exactly what level skills belong to. The stations are also a little to generic.  "Kip progression" doesn't really mean much, even when saying to use resistance bands for assistance.  I do like the gymnastics tag idea.  I have had my gymnasts play gymnastics volleyball (with a beach ball) where they have to do a skill after they hit the ball.

### 6. Drill finder
Sometimes as a coach we can identify what the gymnast needs to do, but finding a drill to work on that piece of the skill can be difficult.  It would be nice if the generator could help with that.

```python
drill_finder = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[
        {"role": "user", 
         "content": """
         I have a gymnast whose jump and arm swing into her back handspring is awesome, 
         however, she leaves her feet behind her and is not kicking them to get over.  
         Provide 3 detailed drills to help with this issue.  Include any equipment needed.
         """}
    ]
)
```

```python
drill_finder_content = drill_finder.choices[0].message.content
print(f"{drill_finder_content}")
```

#### Review of response:
I would use the spotting drill for sure.  The other two would not necessarily be my go to drills for this issue, but they do work for this issue as well. I also like how detailed this response was compared to some of the others.
